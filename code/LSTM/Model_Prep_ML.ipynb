```python
# Leemos los archivos del drive

dataframe = pd.read_csv(os.path.join(MODEL_DATA_PATH, 'df.csv'))

model = load_model(os.path.join(MODEL_PATH, 'lstm_model_keras.keras'))

scaler_X = joblib.load(os.path.join(SCALERS_PATH, 'scaler_X.pkl'))

scaler_y = joblib.load(os.path.join(SCALERS_PATH, 'scaler_y.pkl'))

feature_columns = ['cmcRank', 'circulatingSupply', 'totalSupply', 'price', 'volume24h', 'marketCap', 'percentChange1h', 'percentChange24h',
                'percentChange7d', 'class', 'days_diff', 'name_code', 'days_after_halving']

target_column = 'marketCap_percent_change'
```

```python
# Obtener la lista de tokens únicos
unique_tokens = dataframe['name'].unique().tolist()
```

```python
def predict(name, df):
  # Este es el dataset que retornaremos para verificar la prediccion
  df_return = df[df['name'] == name].head(11)
  # Este es el dataset con el que haremos la prediccion
  df = df[df['name'] == name].head(10)
  # Seleccionar las columnas de features y target
  X = df[feature_columns].to_numpy()
  y = df[target_column].to_numpy().reshape(-1, 1)

  # Aplicar escalado
  X_scaled = scaler_X.transform(X)
  y_scaled = scaler_y.transform(y)

  # Asignar las columnas escaladas al DataFrame
  df[feature_columns] = X_scaled
  df[target_column] = y_scaled

  X_pred = df[feature_columns].to_numpy()
  X_pred = np.reshape(X_pred, (1, X_pred.shape[0], X_pred.shape[1]))

  # Realizamos la predicción
  prediccion = model.predict(X_pred)

  print("Predicción escalada:", prediccion)

  # Re escalamos la prediccion para volver a su valor original
  prediccion_original = scaler_y.inverse_transform(prediccion)

  return prediccion_original[0][0], df_return
```

```python
pred, token_df = predict('GAMEE', dataframe)
```

```python
import plotly.graph_objects as go

def plot_pred_interactive(dataset, pred):
    # Asegúrate de que 'last_updated' esté en formato de fecha
    dataset['last_updated'] = pd.to_datetime(dataset['last_updated'])

    # Seleccionar los primeros 10 registros
    data = dataset.head(10)

    # Obtener la última fecha de los primeros 10 registros y agregar 7 días
    last_date = pd.to_datetime(data['last_updated'].iloc[-1])
    prediction_date = last_date + pd.Timedelta(days=7)

    # Agregar la predicción como el valor 11
    data_pred = data.copy()

    # Crear un diccionario con las mismas columnas que data_pred y llenarlo con NaN para los valores faltantes
    new_row = {col: np.nan for col in data_pred.columns}
    new_row['last_updated'] = prediction_date
    new_row['marketCap_percent_change'] = pred

    data_pred.loc[len(data_pred)] = new_row

    # Crear la figura de Plotly
    fig = go.Figure()

    # Graficar los primeros 10 registros con línea azul
    fig.add_trace(go.Scatter(
        x=data['last_updated'],
        y=data['marketCap_percent_change'],
        mode='markers+lines',
        name='Datos reales',
        line=dict(color='blue'),
        marker=dict(symbol='circle', size=8)
    ))

    # Graficar la predicción con línea roja
    fig.add_trace(go.Scatter(
        x=data_pred['last_updated'].iloc[9:],
        y=data_pred['marketCap_percent_change'].iloc[9:],
        mode='markers+lines',
        name='Predicción',
        line=dict(color='red'),
        marker=dict(symbol='x', size=10)
    ))

    # Añadir título y etiquetas
    fig.update_layout(
        title='Cambio porcentual en Market Cap con Predicción',
        xaxis_title='Fecha',
        yaxis_title='Cambio porcentual en Market Cap',
        xaxis_tickangle=45,
        template='plotly_dark',  # Cambia el tema según tus preferencias
        showlegend=True
    )

    # Convertir a HTML para enviarlo al frontend
    html_str = fig.to_html(full_html=False)

    # Mostrar gráfico interactivo
    fig.show()

    return html_str
```

```python
plot = plot_pred_interactive(token_df, pred)
```
